```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/000/940/original/netflix.csv")
df
```

```python
df.drop("description", axis = 1, inplace = True)
```

```python
df[df.duplicated()]
```

```python
constraint = df["cast"].apply(lambda x: str(x).split(",")).tolist()
df_cast = pd.DataFrame(constraint, index = df["title"])
df_cast = df_cast.stack()
df_cast = pd.DataFrame(df_cast.reset_index())
df_cast= df_cast.rename(columns = {0:"actors"})
df_cast.drop(["level_1"], axis = 1, inplace = True)
```

```python
df_cast
```

```python
df.columns
```

```python
df_final = df_cast.merge(df[['show_id', 'type', 'title', 'director', 'cast', 'country', 'date_added',
       'release_year', 'rating', 'duration', 'listed_in']], on = ["title"], how = "left")
df_final
```

```python
df_final.isnull().sum()
```

```python
df_final["actors"].value_counts()
```

```python
constraint = df["director"].apply(lambda x: str(x).split(",")).tolist()
df_director = pd.DataFrame(constraint, index = df["title"])
df_director = df_director.stack()
df_director = pd.DataFrame(df_director.reset_index())
df_director= df_director.rename(columns = {0:"Director"})
df_director.drop(["level_1"], axis = 1, inplace = True)
```

```python
df_director
```

```python
df_final.columns
```

```python
df_final.drop("cast",axis = 1, inplace = True)
df_final
```

```python
df_final.columns
```

```python
df_final = df_director.merge(df_final[['title', 'actors', 'show_id', 'type', 'director', 'country',
       'date_added', 'release_year', 'rating', 'duration', 'listed_in']], on=["title"], how="left")
```

```python
df_final
```

```python
df_final.drop("director", axis = 1, inplace  = True)
```

```python
df_final
```

```python
constraint = df["country"].apply(lambda x: str(x).split(",")).tolist()
df_country = pd.DataFrame(constraint, index = df["title"])
df_country = df_country.stack()
df_country = pd.DataFrame(df_country.reset_index())
df_country= df_country.rename(columns = {0:"Country"})
df_country.drop(["level_1"], axis = 1, inplace = True)
df_country
```

```python
df_final.columns
```

```python
df_final = df_country.merge(df_final[['title', 'Director', 'actors', 'show_id', 'type', 'country',
       'date_added', 'release_year', 'rating', 'duration', 'listed_in']], on = ["title"], how="left")
df_final
```

```python
df_final.drop("country", axis = 1, inplace = True)
df_final
```

```python
constraint = df["listed_in"].apply(lambda x: str(x).split(",")).tolist()
df_genre = pd.DataFrame(constraint, index = df["title"])
df_genre = df_genre.stack()
df_genre = pd.DataFrame(df_genre.reset_index())
df_genre= df_genre.rename(columns = {0:"genre"})
df_genre.drop(["level_1"], axis = 1, inplace = True)
df_genre
```

```python
df_final.columns
```

```python
df_final = df_genre.merge(df_final[['title', 'Country', 'Director', 'actors', 'show_id', 'type',
       'date_added', 'release_year', 'rating', 'duration', 'listed_in']], on=["title"], how ="left")
df_final
```

```python
df_final.drop("listed_in", axis = 1, inplace =True)
df_final
```

```python
df_final["date_added"] = pd.to_datetime(df_final["date_added"])
df_final["month_added"] = df_final["date_added"].dt.month_name()
df_final["year_added"] = df_final["date_added"].dt.year
df_final["day_of_week"] = df_final["date_added"].dt.day_name()
df_final
```

```python
df_final.isnull().sum()
```

```python
df_final.loc[df_final["duration"].isnull()]
```

```python
df_final.loc[126582,"duration"] = "74 min"
df_final.loc[131648,"duration"] = "84 min"
df_final.loc[131782,"duration"] = "66 min"
```

```python
df_final.isnull().sum()
```

```python
df_final.loc[126582,"rating"] = df_final["rating"].mode()[0]
df_final.loc[131648,"rating"] = df_final["rating"].mode()[0]
df_final.loc[131782,"rating"] = df_final["rating"].mode()[0]
```

```python
df_final.loc[[126582,131648,131782]]
```

```python
df_final["date_added"].fillna(df_final["date_added"].mode()[0], inplace=True)
df_final["year_added"].fillna(df_final["year_added"].mode()[0], inplace=True)
df_final["month_added"].fillna(df_final["month_added"].mode()[0], inplace=True)
df_final["day_of_week"].fillna(df_final["day_of_week"].mode()[0], inplace=True)
df_final["rating"].fillna(df_final["rating"].mode()[0], inplace=True)
```

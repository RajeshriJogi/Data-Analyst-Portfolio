```python
df['Valuation ($B)'] = df['Valuation ($B)'].map(lambda x : x.lstrip('$'))
df['Valuation ($B)'] = df['Valuation ($B)'].astype('str').astype('float')

df['Date Joined'] = pd.to_datetime(df['Date Joined'])
```

```python
df['Industry'] = df['Industry'].map(lambda x: x.strip())
df['City'] = df['City'].map(lambda x: x.strip())
df['Country'] = df['Country'].map(lambda x: x.strip())
df['Company'] = df['Company'].map(lambda x: x.strip())
```

```python
df['Industry'] = df['Industry'].replace(['Artificial intelligence'], 'Artificial Intelligence')
```

```python
df.to_csv('unicorns_basic.csv')
```

```python
df["Investors"] = df["Investors"].replace('.',',')
```

```python
constraint = df["Investors"].apply(lambda x: str(x).split(',')).tolist()
df_investor = pd.DataFrame(constraint, index = df["Company"])
df_investor = df_investor.stack()
df_investor = pd.DataFrame(df_investor.reset_index())
df_investor= df_investor.rename(columns = {0:"Investors_new"})
df_investor.drop(["level_1"], axis = 1, inplace = True)
```

```python
df_investor
```

```python
df_final = df_investor.merge(df[['Company', 'Valuation ($B)', 'Date Joined', 'Country', 'City',
       'Industry', 'Investors']], on = ['Company'], how = 'left')
df_final.drop('Investors', axis = 1, inplace = True)
df_final
```

```python
df_final['Investors_new'] = df_final['Investors_new'].map(lambda x: x.strip())
```

```python
investor_value_count = df_final['Investors_new'].value_counts().sort_index(ascending = True).to_frame()
```
